```python
import os
import kagglehub

path = kagglehub.dataset_download("shahir/protein-data-set")
files = os.listdir(path)

print("Files in the directory:", files)
```

```python
import pandas as pd

pdb_data_no_dups = pd.read_csv(os.path.join(path, files[0]))
pdb_data_seq = pd.read_csv(os.path.join(path, files[1]))

pdb_data_seq.head()
```

```python
pdb_data_seq = pdb_data_seq.drop(columns=['chainId', 'residueCount'])
pdb_data_seq
```

```python
pdb_data_no_dups.head()
```

```python
pdb_data_no_dups = pdb_data_no_dups.drop(columns=['experimentalTechnique', 'residueCount', 'resolution', 'structureMolecularWeight', 'crystallizationMethod', 'crystallizationTempK', 'densityMatthews', 'densityPercentSol', 'pdbxDetails', 'phValue', 'publicationYear'])
pdb_data_no_dups
```

```python
protein_data = pdb_data_no_dups[pdb_data_no_dups['macromoleculeType'] == 'Protein']
protein_data.head()
```

```python
protein_data_seq = pdb_data_seq[pdb_data_seq['macromoleculeType'] == 'Protein']
protein_data_seq.head()
```

```python
print("Length of protein_data:", len(protein_data))
print("Length of protein_data_seq:", len(protein_data_seq))
```

```python
structure_id = '101M'  # Example structure ID

# Find rows in pdb_data_no_dups
rows_no_dups = pdb_data_no_dups[pdb_data_no_dups['structureId'] == structure_id]
print("Rows in pdb_data_no_dups with structureId", structure_id)
rows_no_dups
```

```python
# Find rows in pdb_data_seq
rows_seq = pdb_data_seq[pdb_data_seq['structureId'] == structure_id]
print("Rows in pdb_data_seq with structureId", structure_id)
rows_seq
```

```python
protein_data = protein_data.drop(columns=['macromoleculeType'])
protein_data_seq = protein_data_seq.drop(columns=['macromoleculeType'])

protein_data
```

```python
protein_data_seq
```

```python
merged_protein_data = pd.merge(protein_data_seq, protein_data, on='structureId')
merged_protein_data.head()
```

```python
print("\nNumber of NaN rows in merged_protein_data:")
print(merged_protein_data.isna().sum())
print("\nSize of merged_protein_data:")
print(merged_protein_data.shape)
```

```python
nan_rows = merged_protein_data[merged_protein_data.isna().any(axis=1)]
print(nan_rows)
```

```python
merged_protein_data_cleaned = merged_protein_data.dropna()
merged_protein_data_cleaned
```

```python
print("\nNumber of NaN rows in cleaned merged_protein_data:")
print(merged_protein_data_cleaned.isna().sum())
print("\nSize of cleaned merged_protein_data:")
print(merged_protein_data_cleaned.shape)
```

```python
merged_protein_data_cleaned = merged_protein_data_cleaned.drop(columns=['structureId'])
merged_protein_data_cleaned.head()
```

```python
merged_protein_data_cleaned.to_csv('../data/prototype/merged_protein_data_cleaned.csv', index=False)
```
